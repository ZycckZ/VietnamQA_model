#Huấn luyện mô hình

```python
#tải mmô hình phobert-base để xử lý các trường câu hỏi và ngữ cảnh:
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("KoichiYasuoka/bert-base-vietnamese-ud-goeswith")
```

```python
print(type(tokenizer))
```

```python
def preprocess_function(examples):

#1. Tiền xử lý câu hỏi và token hóa
    questions = [q.strip() for q in examples["question"]]

    inputs = tokenizer( #tokenizer là một bộ token hóa từ một thư viện NLP như Hugging Face Transformers,
                       #nó sẽ mã hóa các câu hỏi (questions) cùng với các đoạn văn bản ngữ cảnh (examples["context"]).

        questions,

        examples["context"],

        max_length=384, # giới hạn độ dài của các chuỗi đầu ra ở 384 token.

        truncation="only_second", #Nếu chuỗi vượt quá max_length, chỉ cắt bớt phần thứ hai (context),
                        #giữ nguyên câu hỏi.

        return_offsets_mapping=True, #Trả về "offset mapping" (ánh xạ offset) để xác định vị trí của
                        #từng token trong văn bản gốc.

        padding="max_length", #thêm padding (đệm) để mọi chuỗi đều có độ dài max_length.

    )

#2. Khởi tạo danh sách vị trí bắt đầu và kết thúc của câu trả lời
    offset_mapping = inputs.pop("offset_mapping") #offset_mapping là một danh sách chứa thông tin về vị trí (offset)
    # của mỗi token trong chuỗi văn bản gốc và được lấy ra từ inputs.

    answers = examples["answers"] #answers chứa các câu trả lời và vị trí bắt đầu của chúng.

    #start_positions và end_positions sẽ lưu vị trí bắt đầu và kết thúc của các câu trả lời trong văn bản được mã hóa.
    start_positions = []

    end_positions = []

#3. Xác định vị trí token bắt đầu và kết thúc của câu trả lời trong đoạn mã hóa
    for i, offset in enumerate(offset_mapping):
        answer = answers[i]

        # Kiểm tra xem answer["answer_start"] có rỗng không
        if not answer["answer_start"]:
            start_positions.append(0)
            end_positions.append(0)
            continue # Skip to the next example if no answer

        #Với mỗi câu trả lời, start_char và end_char xác định vị trí ký tự bắt đầu và kết thúc của câu trả lời trong văn bản gốc.
        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i) #sequence_ids chứa chỉ số của các token thuộc các phần khác nhau của đầu vào
        #(phần câu hỏi và phần ngữ cảnh).

#*** Tìm điểm khởi đầu và kết thúc của "context" ***
#4. Xác định phạm vi của ngữ cảnh trong đoạn mã hóa
  #Đoạn mã này tìm phần mã hóa của context trong chuỗi đầu vào.
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1
        #context_start và context_end đánh dấu vị trí bắt đầu và kết thúc của phần ngữ cảnh trong chuỗi token hóa.

#5. Kiểm tra nếu câu trả lời nằm trong ngữ cảnh
        #Nếu câu trả lời không nằm hoàn toàn trong phạm vi context, gán (0, 0) cho start_positions và end_positions, tức là không
        #có câu trả lời hợp lệ.
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)

#6.Xác định vị trí token bắt đầu và kết thúc của câu trả lời trong ngữ cảnh
        # Nếu không thì đó là vị trí của mã thông báo bắt đầu và kết thúc
        else:
            #Dòng lệnh này duyệt từ context_start đến context_end để tìm token bắt đầu và kết thúc của câu trả lời.
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)   #lưu lại token đánh dấu bắt đầu của câu trả lời.
            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)     #lưu lại token kết thúc của câu trả lời.

#7.Trả về kết quả
    #inputs được bổ sung thêm start_positions và end_positions, trả về kết quả cuối cùng cho quá trình huấn luyện hoặc đánh giá mô hình.
    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
#Để áp dụng chức năng tiền xử lý trên toàn bộ tập dữ liệu, sử dụng chức năng bản đồ Bộ dữ liệu Huggingface.
#Tăng tốc chức năng bản đồ bằng cách đặt batched=True để xử lý nhiều phần tử của tập dữ liệu cùng một lúc.
tokenized_viquad = viquad.map(preprocess_function, batched=True, remove_columns=viquad["train"].column_names)
```

```python
#Tạo một loạt ví dụ bằng cách sử dụng DefaultDataCollator.
#Không giống như các trình đối chiếu dữ liệu khác trong Huggingface Transformers,
#DefaultDataCollator không áp dụng bất kỳ tiền xử lý bổ sung nào, chẳng hạn như phần đệm.
from transformers import DefaultDataCollator

data_collator1 = DefaultDataCollator()
```

```python
from transformers import AutoModelForQuestionAnswering, TrainingArguments, Trainer

model = AutoModelForQuestionAnswering.from_pretrained("KoichiYasuoka/bert-base-vietnamese-ud-goeswith", num_labels = 2)
```

```python
training_args = TrainingArguments( #lớp dùng để cài đặt các tham số cho quá trình huấn luyện.
    #Tham số bắt buộc duy nhất là output_dir chỉ định nơi lưu mô hình(module)
  output_dir="VieModTest", #output_dir="VieModTest": Đường dẫn để lưu các trọng số mô hình, logs, và các
                                   #thông tin khác sau khi huấn luyện xong.
  eval_steps=100, # Tính toán sau mỗi 100 bước
  eval_strategy="epoch",  # Tính toán dựa trên các bước
  learning_rate=2e-5, #Tốc độ học được đặt là 2×10−52×10−5, đây là một giá trị phổ biến khi huấn luyện các mô hình Transformers.

    #Kích thước batch cho tập huấn luyện và tập đánh giá là 16 mẫu mỗi thiết bị.
  per_device_train_batch_size=16,
  per_device_eval_batch_size=16,

  num_train_epochs=3, #Số vòng lặp huấn luyện (epochs) là 3.
  weight_decay=0.01, #Đặt hệ số suy giảm trọng số để tránh overfitting.
  push_to_hub=True #Kết quả của mô hình sẽ được đẩy lên Hugging Face Hub, nếu người dùng có cấu hình thích hợp.
)
#Chuyển đối số huấn luyện cho Người huấn luyện cùng với mô hình, tập dữ liệu,
#trình mã thông báo và trình đối chiếu dữ liệu.
trainer = Trainer(
    model=model, #Mô hình sẽ được huấn luyện (cần phải định nghĩa trước).
    args=training_args, #Các tham số huấn luyện được cấu hình trong training_args.
    train_dataset=tokenized_viquad["train"],  #Tập dữ liệu huấn luyện, ở đây là tập đã được token hóa từ bộ HotpotQA.
    eval_dataset=tokenized_viquad["test"],    #Tập dữ liệu đánh giá, ở đây là tập kiểm thử đã được token hóa từ bộ HotpotQA.
    tokenizer=tokenizer, #Bộ mã hóa (tokenizer) được sử dụng để mã hóa văn bản.
    data_collator=data_collator1, #Hàm ghép nối các mẫu dữ liệu trong batch, giúp đảm bảo dữ liệu đầu vào có cùng độ dài.
)

trainer.train()
```

```python
#Sau khi đào tạo xong, chia sẻ mô hình(module) với Hub bằng phương thức push_to_hub()
trainer.push_to_hub()
```
